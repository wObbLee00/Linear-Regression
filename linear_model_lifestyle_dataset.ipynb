```python
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("https://github.com/ageron/data/raw/main/lifesat/lifesat.csv")
df.head()
```

```python
X=df[["GDP per capita (USD)"]].values
y=df[["Life satisfaction"]].values
print(X,y)
```

```python
df.plot(kind='scatter',x="GDP per capita (USD)",y="Life satisfaction",grid=True)
```

```python
plt.axis([23_500,62_500,4,9])
plt.show()
```

```python
model=LinearRegression()
model.fit(X, y)
```

```python
test=[[37_655.2]]
print(model.predict(test))
```
